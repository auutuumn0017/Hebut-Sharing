### 1. 编写程序，记录十名学生的信息，至少包括姓名、年龄、出生年、数据统计分析课程实验成绩，程序输出前n个学生的年龄平均值、数据统计分析课程实验成绩平均值，n为函数的输入参数。

```python
students_list = [
    {"name": "张三", "age": 20, "birth_year": 2004, "grade": 88},
    {"name": "李四", "age": 21, "birth_year": 2003, "grade": 92},
    {"name": "王五", "age": 19, "birth_year": 2005, "grade": 76},
    {"name": "赵六", "age": 20, "birth_year": 2004, "grade": 95},
    {"name": "孙七", "age": 22, "birth_year": 2002, "grade": 81},
    {"name": "周八", "age": 21, "birth_year": 2003, "grade": 89},
    {"name": "吴九", "age": 19, "birth_year": 2005, "grade": 90},
    {"name": "郑十", "age": 20, "birth_year": 2004, "grade": 78},
    {"name": "陈十一", "age": 21, "birth_year": 2003, "grade": 85},
    {"name": "刘十二", "age": 22, "birth_year": 2002, "grade": 93}
]

def calculation_avg(list,n):
    if n<=0:
        print("n 必须是正整数")
        return (0, 0) 
    
    if n>len(list):
        print(f"n值大于列表长度,将会计算{n}个学生的平均")
    
    total_age = 0
    total_grade = 0
    for student in list[:n]:# student是一个字典
        total_age+=student["age"]
        total_grade+=student["grade"]

    return total_age/n,total_grade/n# 返回元组,可以加()

n_value=5

avg_age,avg_grade=calculation_avg(students_list,n_value)

print(f"前{n_value}个学生的平均年龄: {avg_age:.2f}")
print(f"前{n_value}个学生的平均成绩: {avg_grade:.2f}")
    
        
```

### 2. 找出1至999之间是13的倍数或者前两位数字是13的数字，输出这些数字，并统计有多少个。

```python
def find():
    count=0
    for i in range(1,1000):
        if i>=130 and i<=139 or i%13==0:# and优先级高于or
            # print(f"{i}符合条件")
            count+=1
    print(f"符合条件的数字有{count}个")

find()
```

### 3. 编写成绩转化为绩点的函数，用98，93，89，73，66分别调用函数，生成对应绩点。

```python
def convert(score):
    if score>=90:    
        return 4.0
    if score<60:
        print("成绩不及格")
        return 0
    return 4-(90-score)*0.1

score_list=[98,93,89,73,66]

for i in score_list:
    gpa=convert(i)
    print(f"{i}的绩点为{gpa:.1f}")# 不写.f会默认保留6位小数（计算机无法精确存储0.1）
    
```

### 4. 随机生成两个长为100且服从标准正态分布的向量，然后将两向量所有偶数位的数值对调，输出所有的4个向量（对调前的2个，对调后的2个）。

```python
import numpy as np
vector1=np.random.randn(100)# randn生成服从标准正态分分布的随机数，一般用rand生成[0,1)之间的随机数
vector2=np.random.randn(100)# 形状为(100,)，改成randn(1,100),形状就是（1,100）
print(f"vector1:{vector1[:10]}")# 因为显示不全，所以只取前10个元素展示一下
print(f"vector2:{vector2[:10]}")

# :  [start : stop]	默认 1	选取一个连续的子序列。
# :: [start : stop : step]
temp=vector1[0::2]
vector1[0::2]=vector2[0::2]
vector2[0::2]=temp

print(f"vector1:{vector1[:10]}")
print(f"vector2:{vector2[:10]}")
```

### 5. 已知XYZ+YZZ=532，其中X、Y和Z为数字，编程求出X，Y和Z的值。

```python
def solve():
    for x in range(10):
        for y in range(10):
            for z in range(10):
                if x*100+y*10+z+y*100+z*10+z==532:
                    print(f"x={x},y={y},z={z}")
                    return x,y,z

solve()
```
